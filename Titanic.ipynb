```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_profiling
```

```python
# read the dataset
titanic = pd.read_csv("titanic_data.csv")
titanic.head()
```

Significado de cada coluna:

Survived (Sobreviveu): 0 = Não, 1 = Sim

Pclass (Classe): Classe de ingresso 1 = 1º, 2 = 2º, 3 = 3º

Sex (Sexo): Sexo do passageiro

Age (Idade): Idade em anos

Sibsp: Quantidade de irmãos / cônjuges a bordo do Titanic

Parch: Quantidade de pais / crianças a bordo do Titanic

Ticket (Bilhete): Número do bilhete de embarque

Fare (Tarifa): Tarifa paga pelo Passageiro

Cabin (Cabine): Número de cabine

Embarked (Embarque): Porto de Embarque (C = Cherbourg, Q=Queenstown, S = Southampton)

```python
#function that removes the columns of a specified list
def remove(cols):
    for i in range(0, len(cols)):
        titanic.drop([cols[i]], inplace =True, axis=1)
    return titanic.head()
    
#Vector columns
columns = ['SibSp','Parch', 'Embarked', 'Ticket', 'Cabin']

remove(columns)
```

```python
#Fill NAs values in age column
titanic['Age'].fillna(0, inplace = True)
titanic.head()
```

```python
# create a new column, inverse of "Survived" only to practice the function 'map'
titanic['Not_Suvived'] = titanic['Survived'].map({0:1, 1:0})
titanic.head()
```

# 1.Quantas pessoas sobreviveram ao desastre, segundo este dataset? Quantas não sobreviveram?

```python
print("Sobreviventes: %s" %titanic['Survived'].value_counts()[1])
print("Não sobreviventes: %s" %titanic['Survived'].value_counts()[0])
```

```python
#plot the pie chart
_ = titanic['Survived'].value_counts().plot.pie(colors = ('red','blue'), title = 'Porcentagem dos passageiros sobreviventes', 
                                                 ylabel = '', fontsize = 10, shadow = True, startangle = 90, autopct = '%1.1f%%',
                                                 labels = ('Não sobreviventes', 'Sobreviventes'), figsize = (4,4)
                                                ) 
```

# 2. A classe em que o passageiro encontrava-se foi um fator decisivo para a sobrevivência?

```python
class_survived = titanic[titanic['Survived'] == 1]['Pclass'].value_counts().sort_values(ascending =True)
_ = class_survived.plot.barh(color = 'yellow', figsize = (10,5), alpha = 0.6)
_ = plt.title('Quantidade de sobreviventes por classe \n Total de %s sobreviventes' %class_survived.sum())
_ = plt.ylabel('Classe')
_ = plt.xlabel('Quantidade')
```

```python
#Proporção de passageiros sobreviventes por classe 
prop_class1 =  (class_survived[1] / (titanic[titanic['Pclass'] == 1]['Pclass'].count()))*100 
print('A proporção de passageiros sobreviventes da classe 1 é: %f' %prop_class1)
prop_class2 =  (class_survived[2] / (titanic[titanic['Pclass'] == 2]['Pclass'].count()))*100 
print('A proporção de passageiros sobreviventes da classe 2 é: %f' %prop_class2)
prop_class3 =  (class_survived[3] / (titanic[titanic['Pclass'] == 3]['Pclass'].count()))*100 
print('A proporção de passageiros sobreviventes da classe 3 é: %f' %prop_class3)
```

# 3. Quantidade de homens e mulheres que sobreviveram a tragédia?


```python
#transforma female = 0 e male = 1
f_m_survived = titanic[titanic['Survived']==1]['Sex'].map({'female':0,'male':1}).value_counts()
f_m_survived.plot(kind = 'bar',color = 'red',title = 'Quantidade de sobreviventes por sexo',figsize = (10,5))

plt.xticks(ticks=[0,1], labels = ['female', 'male'], rotation = 0)
plt.xlabel('Sexo')
plt.ylabel('Quantidade')
```

```python
print('Mulheres sobreviventes: %i' %f_m_survived[0])
print('Homens sobreviventes: %i' %f_m_survived[1])
```
